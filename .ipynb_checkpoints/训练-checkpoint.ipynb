```python
import cv2
import numpy as np
import time
import tensorflow as tf

DATA_DIR = 'data/'
variables_save_path = "C:\\Users\\Administrator\\人脸识别\\model-parameters.ckpt"
sess = tf.InteractiveSession()
```

```python
class DataSet(object):
    def __init__(self, images, labels, num_examples):
        self._images = images
        self._labels = labels
        self._epochs_completed = 0  # 完成遍历轮数
        self._index_in_epochs = 0   # 调用next_batch()函数后记住上一次位置
        self._num_examples = num_examples  # 训练样本数

    def next_batch(self, batch_size, fake_data=False, shuffle=True):
        start = self._index_in_epochs

        if self._epochs_completed == 0 and start == 0 and shuffle:
            index0 = np.arange(self._num_examples)
            np.random.shuffle(index0)
            self._images = np.array(self._images)[index0]
            self._labels = np.array(self._labels)[index0]

        if start + batch_size > self._num_examples:
            self._epochs_completed += 1
            rest_num_examples = self._num_examples - start
            images_rest_part = self._images[start:self._num_examples]
            labels_rest_part = self._labels[start:self._num_examples]
            if shuffle:
                index = np.arange(self._num_examples)
                np.random.shuffle(index)
                self._images = self._images[index]
                self._labels = self._labels[index]
            start = 0
            self._index_in_epochs = batch_size - rest_num_examples
            end = self._index_in_epochs
            images_new_part = self._images[start:end]
            labels_new_part = self._labels[start:end]
            return np.concatenate((images_rest_part, images_new_part), axis=0), np.concatenate(
                (labels_rest_part, labels_new_part), axis=0)

        else:
            self._index_in_epochs += batch_size
            end = self._index_in_epochs
            return self._images[start:end], self._labels[start:end]
```

```python
def resize_without_deformation(image, size = (112, 112)):
    height, width, _ = image.shape
    longest_edge = max(height, width)
    top, bottom, left, right = 0, 0, 0, 0
    if height < longest_edge:
        height_diff = longest_edge - height
        top = int(height_diff / 2)
        bottom = height_diff - top
    elif width < longest_edge:
        width_diff = longest_edge - width
        left = int(width_diff / 2)
        right = width_diff - left

    image_with_border = cv2.copyMakeBorder(image, top , bottom, left, right, cv2.BORDER_CONSTANT, value = [0, 0, 0])

    resized_image = cv2.resize(image_with_border, size)

    return resized_image

def read_image(img_num, size = None):
    data_x, data_y = [], []
    for i in range(1,img_num+1):
        try:
            img = cv2.imread(DATA_DIR + 's%d.bmp' % i)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            if size is None:
                size = (112, 112)
            if img.shape != size:
                img = resize_without_deformation(img, size)
            data_x.append(np.asarray(img, dtype = np.float32))
            data_y.append(str(int((i-1)/8.0)))
        except IOError as e:
            print(e)
        except:
            print('Unknown Error!')

    return data_x, data_y
```

```python
def weight_variable(shape):
    initial = tf.truncated_normal(shape, stddev=0.1)
    return tf.Variable(initial)

def basis_variable(shape):
    initial = tf.constant(0.1,shape=shape)
    return tf.Variable(initial)

def conv2d(x,W):
    return tf.nn.conv2d(x, W, strides=[1,1,1,1], padding='SAME')

def max_pool_2x2(x):
    return tf.nn.max_pool(x, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')
```

```python
x = tf.placeholder("float", shape=[None, 112, 112])
y_ = tf.placeholder("float", shape=[None, 41])

W_conv1 = weight_variable([5,5,1,32])
b_conv1 = basis_variable([32])
x_image = tf.reshape(x, [-1,112,112,1])
h_conv1 = tf.nn.relu(conv2d(x_image, W_conv1) + b_conv1)
h_pool1 = max_pool_2x2(h_conv1)

W_conv2 = weight_variable([5,5,32,32])
b_conv2 = basis_variable([32])
h_conv2 = tf.nn.relu(conv2d(h_pool1, W_conv2) + b_conv2)
h_pool2 = max_pool_2x2(h_conv2)

W_conv3 = weight_variable([5,5,32,32])
b_conv3 = basis_variable([32])
h_conv3 = tf.nn.relu(conv2d(h_pool2, W_conv3) + b_conv3)
h_pool3 = max_pool_2x2(h_conv3)

W_conv4 = weight_variable([5,5,32,64])
b_conv4 = basis_variable([64])
h_conv4 = tf.nn.relu(conv2d(h_pool3, W_conv4) + b_conv4)
h_pool4 = max_pool_2x2(h_conv4)

W_fc1 = weight_variable([7*7*64, 2048])
b_fc1 = basis_variable([2048])
h_pool4_flat = tf.reshape(h_pool4, [-1, 7*7*64])
h_fc1 = tf.nn.relu(tf.matmul(h_pool4_flat, W_fc1) + b_fc1)

keep_prob = tf.placeholder(tf.float32)
h_fc1_drop = tf.nn.dropout(h_fc1, keep_prob)

W_fc2 = weight_variable([2048, 41])
b_fc2 = basis_variable([41])
y_conv = tf.nn.softmax(tf.matmul(h_fc1_drop, W_fc2) + b_fc2)

cross_entropy =tf.reduce_mean(-tf.reduce_sum(y_*tf.log(tf.clip_by_value(y_conv, 1e-10, 1.0)), reduction_indices=[1]))
train_step = tf.train.AdamOptimizer(1e-5).minimize(cross_entropy)

correct_prediction = tf.equal(tf.argmax(y_conv,1),tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
def train():
    starttime = time.time()
    tf.global_variables_initializer().run()
    saver = tf.train.Saver()
    IMAGE_SIZE = 112
    Classes = 41
    raw_images, raw_labels = read_image(328, size=(IMAGE_SIZE, IMAGE_SIZE))
    raw_images, raw_labels = np.asarray(raw_images, dtype = np.float32), np.asarray(raw_labels, dtype = np.int32)
    cv2.imshow('', raw_images[0])
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    raw_images /= 255.0
    cv2.imshow('', raw_images[0])
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    one_hot_labels = tf.one_hot(raw_labels, Classes).eval()
    faces = DataSet(raw_images, one_hot_labels, 328)
    for step in range(10000):
        img_batch, lab_batch = faces.next_batch(41)
        print(lab_batch[0])
        cv2.imshow('', img_batch[0])
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        if step%50 == 0:
            train_accuracy = accuracy.eval(feed_dict={x:img_batch, y_: lab_batch, keep_prob: 1.0})
            print("step %d, training accuracy %g"%(step,train_accuracy))
        train_step.run(feed_dict={x:img_batch, y_: lab_batch, keep_prob: 0.5})
    endtime = time.time()
    saver.save(sess, variables_save_path)
    print("traintime: %dh%dm"%((endtime-starttime)//3600, (endtime-starttime)//60%60))

    
if __name__=='__main__':
        train()
    
```
